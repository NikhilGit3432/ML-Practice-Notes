## Data Interpolation

### Linear Interpolation
Connects two known data points with a straight line and estimates values within the line.

### Polynomial Interpolation
Fits a polynomial curve through all known points. The curve is used to estimate unknown points.

### Cubic Interpolation
Uses cubic polynomials to estimate values, providing smoother results than linear interpolation.

# Linear Interpolation

```python
## Make a dataset which follows a Linear Function 
import numpy as np
import pandas as pd

x =np.arange(1,10,dtype=int)
print(x)
y = []
for i in x:
    i = i
    y.append(i)
print(y)    
```

```python
## Put x and y as variables in a dataframe
df = pd.DataFrame(x,columns =["F"])
df["T"] = y
df
```

```python
## Plot a Line Chart  for the df
import matplotlib.pyplot as plt
```

```python
# Plot a line chart
plt.plot(df["F"], df["T"], marker='o', linestyle='-', color='b', label='y = x^2')
plt.title("Line Chart of y = x^2")
plt.xlabel("F (x values)")
plt.ylabel("T (y values)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
## Linear Iterpolaton
## MAke new Data points for new versiion of x such that the data points here in this new variable of x are linearly spaced
x_new = np.linspace(1,9,20)
x_new
```

```python
## Make a new y variae whic h shall have interpolated values corresponding to x_new variable 
y_new = np.interp(x_new,x,y)
y_new
```

```python
# Plot a scatter plot for x_new and y_new
plt.scatter(x_new, y_new, color='red', label='Interpolated Points')

# Add the original line for reference
plt.plot(x, y, color='blue', linestyle='-', marker='o', label='Original Data')

# Title and labels
plt.title("Scatter Plot with Linear Interpolation")
plt.xlabel("x_new (Interpolated x values)")
plt.ylabel("y_new (Interpolated y values)")
plt.legend()
plt.grid(True)
plt.show()
```

# Cubic Interpolation

```python
# since we have already made x variable we can go with making a y variable which will have cubic relation with the same x variable 
y3 = []
for i in x:
    i=i**3
    y3.append(i)

y3
    
```

```python
df_3 = pd.DataFrame(x, columns = ["F3"])
df_3["T3"] = y3
```

```python
# Plot a line chart
plt.plot(df_3["F3"], df_3["T3"], marker='o', linestyle='-', color='green', label='y = x^3')
plt.title("Line Chart of y = x^3")
plt.xlabel("F3 (x values)")
plt.ylabel("T3 (y values)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# make a new y_3 variable which should have a cubically interpolated values 
# for that first we will have to make a new X_3 variable which should have equally spaced data ponts in the range as x variable 
x3_new = np.linspace(1,9,20)
x3_new

# importing requird interpolation library
from scipy.interpolate import interp1d


# make a function which will carry cubic interpolation 
cubic_f = interp1d(x,y3,kind="cubic")


# Now make interpolated y3_new variable
y3_new = cubic_f(x3_new)

y3_new
```

```python
plt.scatter(x3_new,y3_new)
```

# Polynomial interpolation

```python
# Since we have already made a x variable, we can now make a y4 varable which should have polynomial fucntion with x variable 
y4 = np.random.randint(1, 10, size=x.shape)
print("Random y values (y4):", y4)
```

```python
# Plot the original scattered data points
plt.scatter(x, y4, color='blue', label='Original Data')
```

```python
# Polynomial Interpolation
# Create a new x variable (x4_new) with 100 equally spaced points between 1 and 9
x4_new = np.linspace(1, 9, num=100)
print("Interpolated x values (x4_new):", x4_new)
```

```python
# Fit a polynomial of degree 2 to the original data points (x, y4)
# This will generate the coefficients for the polynomial
polynomial_coefficients = np.polyfit(x, y4, 2)

# Generate the corresponding y values for the interpolated x values using the polynomial
y4_new = np.polyval(polynomial_coefficients, x4_new)
print("Interpolated y values (y4_new):", y4_new)
```

```python
# Plot the original scattered data points
plt.scatter(x, y4, color='blue', label='Original Data', marker='o', linestyle='dotted')

# Plot the interpolated polynomial data points
plt.plot(x4_new, y4_new, color='red', label='Polynomial Interpolation', linewidth=2)

# Add proper labels and title
plt.title('Polynomial Interpolation Example', fontsize=16)
plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values', fontsize=12)

# Add grid, legend, and show plot
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
